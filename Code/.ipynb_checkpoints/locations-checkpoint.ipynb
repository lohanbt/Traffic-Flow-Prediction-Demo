```python
import numpy as np
import pandas as pd
import re
import networkx as nx
import matplotlib.pyplot as plt
import math
```

```python
columns = ["SCATS Number", "Location", "NB_LATITUDE", "NB_LONGITUDE"]
df = pd.read_csv("data.csv", skiprows=1, usecols=columns, encoding='utf-8').fillna(0)
df=df.drop_duplicates()
print(df.head())
```

```python
def haversine(coord1, coord2):
    # Unpack coordinates
    lat1, lon1 = coord1
    lat2, lon2 = coord2
    
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2
    c = 2 * math.asin(math.sqrt(a))
    
    # Radius of Earth in kilometers (use 3956 for miles)
    r = 6371.0
    
    # Calculate the distance
    distance = c * r
    return distance

def travel_time(coord1, coord2, speed_kmh):
    # Calculate the distance using the haversine function
    distance = haversine(coord1, coord2)
    
    # Calculate travel time
    time_hours = distance / speed_kmh
    return time_hours
```

```python
locations = df["Location"].tolist()
scats = df["SCATS Number"].tolist()
coordinates = df[["NB_LATITUDE", "NB_LONGITUDE"]].values
coordinates = [tuple(row) for row in coordinates]

def extract(data):
  pattern = re.compile(r"(.+)\s([SNEW]+)\s(?:[Oo][Ff])\s(.+)")
  edges = [pattern.match(entry).groups() for entry in data if pattern.match(entry)]
  return edges

edges = extract(locations)
print(edges)
print(len(locations))

G = nx.DiGraph()
unique_scats = set(scats)
for unique_scat in unique_scats:
  G.add_node(unique_scat)

# get unique roads
roads_0 = [t[0] for t in edges]
roads_2 = [t[2] for t in edges]
roads = roads_0 + roads_2
roads = set(roads)

roads_intersects = {road: {direction: [] for direction in ['NS', 'WE', 'NWSE', 'SWNE']} for road in roads}

for i in range(len(edges)):
  edge = edges[i]
  road_0 = edge[0]
  road_2 = edge[2]
  direction = edge[1]
  coordinate = coordinates[i]
  scat = scats[i]
  if direction == 'N' or direction == 'S':
    roads_intersects[road_0]['WE'].append((road_2, coordinate, scat))
    roads_intersects[road_2]['NS'].append((road_0, coordinate, scat))
  elif direction == 'W' or direction == 'E':
    roads_intersects[road_0]['NS'].append((road_2, coordinate, scat))
    roads_intersects[road_2]['WE'].append((road_0, coordinate, scat))
  elif direction == 'NW' or direction == 'SE':
    roads_intersects[road_0]['SWNE'].append((road_2, coordinate, scat))
    roads_intersects[road_2]['NWSE'].append((road_0, coordinate, scat))
  elif direction == 'NE' or direction == 'SW':
    roads_intersects[road_0]['NWSE'].append((road_2, coordinate, scat))
    roads_intersects[road_2]['SWNE'].append((road_0, coordinate, scat))


for road in roads_intersects:
  for direction in roads_intersects[road]:
    if len(roads_intersects[road][direction]) == 0 or len(roads_intersects[road][direction]) == 1:
      continue
    if direction == 'NS':
      roads_intersects[road][direction].sort(key=lambda x: x[1][1])
    else:
      roads_intersects[road][direction].sort(key=lambda x: x[1][0])
    for i in range(len(roads_intersects[road][direction])-1):
      distance = travel_time(roads_intersects[road][direction][i][1], roads_intersects[road][direction][i+1][1], 60)
      G.add_edge(roads_intersects[road][direction][i][2], roads_intersects[road][direction][i+1][2], weight=distance)
      G.add_edge(roads_intersects[road][direction][i+1][2], roads_intersects[road][direction][i][2], weight=distance)
```

```python
nx.write_gml(G, "map/graph.gml")
```

```python
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=50, font_size=15, font_color='black')

# Show the plot
plt.show()
```

```python
shortest_path = nx.all_simple_paths(G, source=970, target=4264)
for i, path in enumerate(shortest_path, start=1):
    print(f"Path {i}: {path}")
```
